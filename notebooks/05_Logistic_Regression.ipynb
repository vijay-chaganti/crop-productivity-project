# 05 - Logistic Regression Model
Crop Productivity Prediction Project

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
%matplotlib inline

# Load fresh data (we need Season column)
data_agri = pd.read_csv('../data/Agriculture-Data.csv')
data_agri.head()
```

```python
# Create dummy variables for Weather and Season
Weather = pd.get_dummies(data_agri['Weather'])
Season = pd.get_dummies(data_agri['Season'])

# Concat with main dataframe
data_agri = pd.concat([data_agri, Weather, Season], axis=1)

# Drop original text columns
data_agri.drop(['Season', 'Crop', 'Weather'], axis=1, inplace=True)
data_agri.head()
```

```python
# Features and Target
X = data_agri.drop('Kharif', axis=1)
y = data_agri['Kharif']

print("Features shape:", X.shape)
print("Target shape:", y.shape)
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# Handle missing values
X_train = X_train.fillna(X_train.mean())
X_test = X_test.fillna(X_test.mean())

print("Training size:", X_train.shape)
print("Testing size:", X_test.shape)
```

```python
# Train Logistic Regression Model
logmodel = LogisticRegression()
logmodel.fit(X_train, y_train)

print("Model trained successfully!")
```

```python
# Predictions
pred = logmodel.predict(X_test)

print("Logistic Regression Results:")
print(confusion_matrix(y_test, pred))
print("\n")
print(classification_report(y_test, pred))
```

```python
# Residual Plot
sns.histplot((y_test.astype(int) - pred.astype(int)), bins=20, kde=True)
plt.title("Residual Distribution - Logistic Regression")
plt.xlabel("Residuals")
plt.show()
```
